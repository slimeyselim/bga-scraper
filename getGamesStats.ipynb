```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from IPython.core.display import display, HTML
import csv

url = 'https://boardgamearena.com/gamelist?allGames='
classGameLink = 'bga-game-item'

csv_file = 'gameLinks.csv'

# Path to the ChromeDriver executable
driver_path = 'C:\Google\chromedriver-win64\chromedriver-win64\chromedriver.exe'  # Replace with the actual path to ChromeDriver

# Set up the WebDriver
service = Service(driver_path)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run Chrome in headless mode
driver = webdriver.Chrome(service=service, options=options)

driver.get(url)

hrefs = []

try:
    WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
    rendered_html = driver.page_source
    
    soup = BeautifulSoup(rendered_html, 'html.parser')
    
    for link in soup.find_all('a', class_=classGameLink):
        hrefs.append(link.get('href'))
    
finally:
    # Close the WebDriver
    driver.quit()
    
print(hrefs)
print(len(hrefs))    


with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows([[href] for href in hrefs])

hrefs
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import csv
import time
from datetime import datetime

url = 'https://boardgamearena.com/'
classGameLink = 'Number of games played'

csv_file = 'gameLinks.csv'
csv_output = 'gamesPlayed.csv'

# Path to the ChromeDriver executable
driver_path = 'C:\Google\chromedriver-win64\chromedriver-win64\chromedriver.exe'  # Replace with the actual path to ChromeDriver

# Set up the WebDriver
service = Service(driver_path)
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run Chrome in headless mode
driver = webdriver.Chrome(service=service, options=options)

# Function to extract HTML content from a webpage given its URL
def extract_html(url):
    driver.get(url)
    # You can add code here to extract specific elements or perform actions on the webpage
    html_content = driver.page_source
    return html_content


# Open the CSV file and iterate through each href
with open(csv_file, newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row[0])
        href = row[0]  # Assuming the href is in the first column
        WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
        currentUrl = f"{url}{href}"  # Replace example.com with your base URL
        print(currentUrl)
        html_content = extract_html(currentUrl)
        # Do something with the html_content, such as saving it to a file
        #print(html_content)
        soup = BeautifulSoup(html_content, 'html.parser')

        for info in soup.find_all('div',class_='bga-link'):
            #print(info.get_text())

            try:
                numberOfGames = int(info.get_text().replace(',',''))
                print(href)
                print(numberOfGames)
                with open(csv_output, 'a', newline='') as file:
                    writer = csv.writer(file)
                    writer.writerow([href,numberOfGames,datetime.now().strftime('%Y-%m-%d')])
            except ValueError as e:
                pass
        time.sleep(3)
driver.quit()
```
